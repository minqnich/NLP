```python
# Nichakamol Sanghong 642115015
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
import re

df = pd.read_csv('spam.csv', encoding='ISO-8859-1')

if 'v1' in df.columns and 'v2' in df.columns:

    df['v2'] = df['v2'].str.strip()

    df['v2'] = df['v2'].apply(lambda x: ' '.join(re.findall(r'\b\w+\b', str(x))))

    df['length'] = df['v2'].apply(lambda x: len(x.split()))

    df['text1'] = df['v2']

    label_encoder = LabelEncoder()
    df['label'] = label_encoder.fit_transform(df['v1'])

    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(df['text1'])

    feature_names = vectorizer.get_feature_names_out()

    transformed_sample = X[0].toarray().flatten()

    result_df = pd.DataFrame({'Word': feature_names, 'Count': transformed_sample})
    result_df = result_df.sort_values(by='Count', ascending=False)

    text1_df = df[['v1', 'text1', 'length']].copy()
    text1_df.index = range(len(text1_df))

    print("original dataframe:")
    print(text1_df)

    df['text2'] = df['text1'].apply(lambda x: ' '.join(re.findall(r'\b\w+\b', str(x)))) 
    df['length'] = df['text2'].apply(lambda x: len(x.split())) 

    target_max_length_text2 = 26

    df['text2'] = df.apply(lambda row: ' '.join(row['text2'].split()[:target_max_length_text2]), axis=1)
    df['length'] = df['text2'].apply(lambda x: len(x.split())) 

    processed_df = df[['v1', 'text2', 'length']].copy()
    processed_df.index = range(len(processed_df))

    print("\nprocessed DataFrame:")
    print(processed_df)

else:
    print("Columns 'v1' and 'v2' not found in the DataFrame. Please check the actual column names.")






```
